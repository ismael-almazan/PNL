```python
# Librerias requeridas
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords
```

## Ejercicio 1: Utiliza el texto crudo (sin procesar)
*   Crear un CountVectorizer y un TfidfVectorizer.
*	Aplícalos al corpus y transformar los textos.
*	Mostrar:
    *	Dimensiones de la matriz resultante.
    *	Primeros 10 tokens y sus índices.
*	Comparar las diferencias entre ambas matrices (CountVectorizer y TfidfVectorizer), mencionando de manera general las diferencias observadas.

*La principal diferencia entre las matrices generadas por CountVectorizer y TfidfVectorizer es que la primera representa el número de veces que aparece cada palabra en los documentos, por lo que los valores son frecuencias enteras. En cambio, TF-IDF asigna pesos a las palabras considerando su frecuencia en el documento y su importancia en todo el corpus, por lo que los valores son decimales y las palabras muy comunes tienen menor peso.*


```python
#  CounVectorizer
def fcountvectorizer(docs):

    vectorizer = CountVectorizer(lowercase=False)

    X = vectorizer.fit_transform(docs)

    # DataFrame
    bow_df = pd.DataFrame(
        X.toarray(),
        columns=vectorizer.get_feature_names_out()
    )


    return bow_df

bow_df = fcountvectorizer(corpus_es)
#Dimension de la matriz
print("Dimensión de la matriz: ",bow_df.shape)
#Primeros 10 tokens y sus índices
print("\n CounVectorizer (frecuencia)")
display(bow_df.head(10))




#  TfidfVectorizer
def ftfidfvectorizer(docs):
    vectorizer = TfidfVectorizer(lowercase=False)

    X = vectorizer.fit_transform(docs)

    # DataFrame
    tfidf_df = pd.DataFrame(
        X.toarray(),
        columns=vectorizer.get_feature_names_out()
    )

    return tfidf_df

tfidf_df = ftfidfvectorizer(corpus_es)
#Dimension de la matriz
print("\n Dimensión de la matriz: ",tfidf_df.shape)
#Primeros 10 tokens y sus índices
print("\n TfidfVectorizer (frecuencia)")
display(tfidf_df.head(10))
```

##  Ejercicio 2: Preprocesa el texto eliminando los acentos y stopwords, así como convirtiéndolo a minúsculas.
*	Realiza las mismas actividades que en el ejercicio anterior.
*	Compara las matrices del ejercicio anterior con las matrices que obtuviste en este ejercicio, mencionando de manera general las diferencias observadas.

 *Después del preprocesamiento del texto (convertir a minúsculas, eliminar acentos y quitar stopwords), el vocabulario se reduce y las matrices tienen menos columnas, ya que se eliminan palabras que aportan poca informacion. Además, las palabras se estandarizan, lo que permite que la representación del texto sea más limpia y enfocada en los términos más relevantes.*

```python
# Para eliminar lss stopwords se puede hacer desde la lista de stop ya que contiene la libreria de NLTK, ya que CountVectorizer solo funciona para ingles
nltk.download('stopwords')
stopwords_es = stopwords.words('spanish')



#  CounVectorizer
def fcountvectorizer(docs):

    vectorizer = CountVectorizer(lowercase=True, strip_accents='unicode', stop_words=stopwords_es)

    X = vectorizer.fit_transform(docs)

    # DataFrame
    bow_df = pd.DataFrame(
        X.toarray(),
        columns=vectorizer.get_feature_names_out()
    )

    return bow_df

bow_df = fcountvectorizer(corpus_es)
#Dimension de la matriz
print("Dimensión de la matriz: ",bow_df.shape)
#Primeros 10 tokens y sus índices
print("\n CounVectorizer (frecuencia)")
display(bow_df.head(10))




#  TfidfVectorizer
def ftfidfvectorizer(docs):
    vectorizer = TfidfVectorizer(lowercase=True, strip_accents='unicode', stop_words=stopwords_es)

    X = vectorizer.fit_transform(docs)

    # DataFrame
    tfidf_df = pd.DataFrame(
        X.toarray(),
        columns=vectorizer.get_feature_names_out()
    )

    return tfidf_df

tfidf_df = ftfidfvectorizer(corpus_es)
#Dimension de la matriz
print("\n Dimensión de la matriz: ",tfidf_df.shape)
#Primeros 10 tokens y sus índices
print("\n TfidfVectorizer (frecuencia)")
display(tfidf_df.head(10))
```
